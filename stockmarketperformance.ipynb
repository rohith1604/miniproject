```python
import pandas as pd
import yfinance as yf
from datetime import datetime

start_date = datetime.now() - pd.DateOffset(months=3)
end_date = datetime.now()

tickers = ['AAPL', 'MSFT', 'NFLX', 'GOOG', 'TSLA', 'NVDA', 'AMZN']

df_list = []

for ticker in tickers:
    data = yf.download(ticker, start=start_date, end=end_date)
    df_list.append(data)

df = pd.concat(df_list, keys=tickers, names=['Ticker', 'Date'])
print(df.head())
```

```python
df = df.reset_index()
print(df.head())
```

```python
import plotly.express as px
fig = px.line(df, x='Date',
              y='Close',
              color='Ticker',
              title="Stock Market Performance for the Last 3 Months")
fig.show()
```

```python
fig = px.area(df, x='Date', y='Close', color='Ticker',
              facet_col='Ticker',
              labels={'Date':'Date', 'Close':'Closing Price', 'Ticker':'Company'},
              title='Stock Prices for Apple, Microsoft, Netflix, and Google')
fig.show()
```

```python
df['MA10'] = df.groupby('Ticker')['Close'].rolling(window=10).mean().reset_index(0, drop=True)
df['MA20'] = df.groupby('Ticker')['Close'].rolling(window=20).mean().reset_index(0, drop=True)
df['MA20'] = df.groupby('Ticker')['Close'].rolling(window=20).mean().reset_index(0, drop=True)

for ticker, group in df.groupby('Ticker'):
    print(f'Moving Averages for {ticker}')
    print(group[['MA10', 'MA20']])
```

```python
for ticker, group in df.groupby('Ticker'):
    fig = px.line(group, x='Date', y=['Close', 'MA10', 'MA20'],
                  title=f"{ticker} Moving Averages")
    fig.show()
```

```python
df['Volatility'] = df.groupby('Ticker')['Close'].pct_change().rolling(window=10).std().reset_index(0, drop=True)
fig = px.line(df, x='Date', y='Volatility',
              color='Ticker',
              title='Volatility of All Companies')
fig.show()
```

```python
# create a DataFrame with the stock prices of Apple and Microsoft
apple = df.loc[df['Ticker'] == 'AAPL', ['Date', 'Close']].rename(columns={'Close': 'AAPL'})
microsoft = df.loc[df['Ticker'] == 'MSFT', ['Date', 'Close']].rename(columns={'Close': 'MSFT'})
df_corr = pd.merge(apple, microsoft, on='Date')

# create a scatter plot to visualize the correlation
fig = px.scatter(df_corr, x='AAPL', y='MSFT',
                 trendline='ols',
                 title='Correlation between Apple and Microsoft')
fig.show()
```

```python
import pandas as pd
import yfinance as yf
from datetime import datetime

start_date = datetime.now() - pd.DateOffset(months=3)
end_date = datetime.now()

tickers = ['AAPL', 'MSFT', 'NFLX', 'GOOG', 'TSLA', 'NVDA', 'AMZN']

df_list = []

for ticker in tickers:
    data = yf.download(ticker, start=start_date, end=end_date)
    df_list.append(data)

df = pd.concat(df_list, keys=tickers, names=['Ticker', 'Date'])
print(df.head())
```

```python
# Calculate daily returns for each stock
df['Daily Return'] = df['Adj Close'].groupby('Ticker').pct_change()

# Drop NaN values resulting from the percentage change
df = df.dropna()

print(df.head())
```

```python
# Calculate cumulative returns for each stock
df['Cumulative Return'] = (1 + df['Daily Return']).groupby('Ticker').cumprod()
print(df.head())
```

```python
import plotly.express as px

# Resetting the index to make 'Date' a regular column
df_reset = df.reset_index()

# Plot closing prices for each stock using Plotly
fig = px.line(df_reset, x='Date', y='Adj Close', color='Ticker',
              labels={'Adj Close': 'Adjusted Close Price'},
              title='Closing Prices for Selected Stocks',
              template='plotly_dark')

# Show the plot
fig.show()
```

```python
import plotly.express as px

# Calculate 50-day and 200-day moving averages for a specific stock (e.g., AAPL)
stock_symbol = 'AAPL'

# Create a new DataFrame with the moving averages
ma_df = df.loc[stock_symbol].copy()
ma_df['50-Day MA'] = ma_df['Adj Close'].rolling(window=50).mean()
ma_df['200-Day MA'] = ma_df['Adj Close'].rolling(window=200).mean()

# Plot using Plotly Express
fig = px.line(ma_df, x=ma_df.index, y=['Adj Close', '50-Day MA', '200-Day MA'],
              labels={'value': 'Adjusted Close Price'},
              title=f'{stock_symbol} Stock Price with Moving Averages',
              template='plotly_dark')

# Show the plot
fig.show()
```

```python
import plotly.express as px

# Calculate rolling volatility for a specific stock (e.g., TSLA)
stock_symbol = 'TSLA'

# Create a new DataFrame with the rolling volatility
volatility_df = df.loc[stock_symbol].copy()
volatility_df['Volatility'] = volatility_df['Daily Return'].rolling(window=20).std()

# Plot using Plotly Express
fig = px.line(volatility_df, x=volatility_df.index, y='Volatility',
              labels={'Volatility': 'Volatility (20-Day Rolling Standard Deviation)'},
              title=f'{stock_symbol} Daily Return Volatility',
              template='plotly_dark')

# Show the plot
fig.show()
```

```python
import plotly.express as px

# Calculate 50-day and 200-day moving averages and trading signals
short_window = 50
long_window = 200

df['Short_MA'] = df.groupby('Ticker')['Adj Close'].rolling(window=short_window).mean().reset_index(level=0, drop=True)
df['Long_MA'] = df.groupby('Ticker')['Adj Close'].rolling(window=long_window).mean().reset_index(level=0, drop=True)

df['Signal'] = 0
df['Signal'][df['Short_MA'] > df['Long_MA']] = 1  # Buy signal
df['Signal'][df['Short_MA'] < df['Long_MA']] = -1  # Sell signal

# Plotting the signals using Plotly Express
fig = px.line(df.reset_index(), x='Date', y=['Short_MA', 'Long_MA'],
              labels={'value': 'Moving Average'},
              title='Moving Average Trading Signals',
              template='plotly_dark')

# Add scatter plot for signals
fig.add_scatter(x=df.reset_index()['Date'], y=df['Adj Close'].values, mode='markers', marker=dict(size=8, color=df['Signal'], colorscale='viridis'),
                name='Signals')

# Show the plot
fig.show()
```

```python
import plotly.express as px

# Calculate annualized mean and standard deviation of daily returns
annualized_mean_return = df.groupby('Ticker')['Daily Return'].mean() * 252
annualized_std_dev = df.groupby('Ticker')['Daily Return'].std() * (252 ** 0.5)

# Create a scatter plot using Plotly Express
fig = px.scatter(x=annualized_std_dev, y=annualized_mean_return,
                 labels={'x': 'Annualized Standard Deviation (Risk)', 'y': 'Annualized Mean Return'},
                 title='Risk vs Return for Selected Stocks',
                 text=df.groupby('Ticker').last().index,
                 template='plotly_dark',
                 hover_name=df.groupby('Ticker').last().index)

# Show the plot
fig.show()
```

```python
import pandas as pd
import yfinance as yf
from datetime import datetime

# Download historical stock data
start_date = datetime.now() - pd.DateOffset(months=3)
end_date = datetime.now()

tickers = ['AAPL', 'MSFT', 'NFLX', 'GOOG', 'TSLA', 'NVDA', 'AMZN']

df_list = []

for ticker in tickers:
    data = yf.download(ticker, start=start_date, end=end_date)
    df_list.append(data)

df = pd.concat(df_list, keys=tickers, names=['Ticker', 'Date'])

# Calculate daily returns
df['Daily Return'] = df['Adj Close'].groupby('Ticker').pct_change()

# Identify periods of profit (positive returns) and loss (negative returns)
df['Profit'] = df['Daily Return'] > 0
df['Loss'] = df['Daily Return'] < 0

# Calculate the percentage of profit and loss days
profit_loss_percentage = df.groupby(['Ticker', 'Profit', 'Loss']).size() / df.groupby('Ticker').size() * 100
profit_loss_percentage = profit_loss_percentage.unstack(level=['Profit', 'Loss'], fill_value=0)

# Print the result
print("Percentage of Profit and Loss Days:")
print(profit_loss_percentage)
```
